```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.violinplot(x = df['total_bill'])
```

```python
sns.violinplot(x = 'sex', y = "tip", data = df)
```

```python
# hue and palette
sns.violinplot(x = 'sex', y = "tip", data = df, hue="day", palette = "viridis")
```

```python
# horizonatal and increse line width
sns.violinplot(y = 'sex', x = "tip", data = df, hue="day", palette = "viridis", linewidth=3)
```

```python
# split to combine both
sns.violinplot(x = 'day', y = "tip", data = df, hue="sex", split=True)
```
