```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.histplot(x = "total_bill", data = df)
```

```python
 sns.histplot(x = "total_bill", data = df, hue="sex", palette = "spring", kde = "true")
plt.show()
```

```python
sns.histplot(x = "total_bill", data = df, hue="sex", edgecolor="red", linestyle="--")
plt.show()
```

```python
# color
sns.histplot(x = "total_bill", data = df, color="hotpink")
plt.show()
```
