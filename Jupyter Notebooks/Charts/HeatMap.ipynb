```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = np.random.randint(low = 1, high = 100, size = (10,10))
print(data)
```

```python
sns.heatmap(data = data)
```

```python
sns.heatmap(data = data, cmap = "GnBu")
```

```python
sns.heatmap(data = data, cmap = "GnBu", alpha = 0.5)
```

```python
# add line color
sns.heatmap(data = data, linecolor="yellow", linewidths="4")
```

```python
# show numbers in box
sns.heatmap(data = data, cmap = "GnBu", annot = True)
```

```python
# remove color bar
sns.heatmap(data = data, cmap = "GnBu", cbar=False, annot =True)
```

```python
# remove tick lables
sns.heatmap(data = data, cmap = "GnBu", xticklabels=False, yticklabels=False)
```
