```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.barplot(x = "who", y = "fare", data=df)
plt.show()
```

```python
# coloring
sns.barplot(x = "who", y = "fare", data=df, color="magenta")
plt.show()
```

```python
# hue
sns.barplot(x = "who", y = "fare", data=df, hue="sex")
plt.show()
```

```python
# palatte
sns.barplot(x = "sex", y = "fare", data=df, hue="who", palette="spring")
plt.show()
```

```python
 # saturation
sns.barplot(x = "who", y = "fare", data=df, hue="sex", palette="spring", saturation=0.2)
plt.show()
```

```python
# horizontal bar and ci for remove median line
sns.barplot(x = "fare", y = "who", data=df, ci=False)
plt.show()
```

```python
# display mean

from numpy import mean
sns.barplot(x = "fare", y = "who", data=df, estimator=mean)
plt.show()
```

```python
# display mean

from numpy import median
sns.barplot(x = "fare", y = "who", data=df, estimator=median)
plt.show()
```

```python
#facecolor take 4 paramerter - rgba
sns.barplot(x = "fare", y = "who", data=df, estimator=median, facecolor=[1,1,1,0], edgecolor="0.2")
plt.show()
```

```python
#change order
sns.barplot(x = "who", y = "fare", data=df, estimator=median, order=["child", "womam", "man"])
plt.show()
```
