```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.boxplot(x = "tip", data = df)
# first left side is called minimun, than where box start is called first qurtail, box end is called third qurtial, center line is median and final line is maximum value
```

```python
sns.boxplot(x = "sex", y = "tip", data = df)
```

```python
sns.boxplot(x = "sex", y = "tip", data = df, hue = "smoker")
```

```python
# color only support when we work without hue or with hue we can use pallete
sns.boxplot(x = "sex", y = "tip", data = df, hue = "smoker", palette = "GnBu")
```

```python
sns.boxplot(x = "sex", y = "tip", data = df, color = "hotpink")
```

```python
# Orientation
sns.boxplot(data = df, orient="h")
```

```python
# add grid lines
sns.set(style = "whitegrid")
sns.boxplot(x = "sex", y = "tip", data = df, hue = "smoker")
```
