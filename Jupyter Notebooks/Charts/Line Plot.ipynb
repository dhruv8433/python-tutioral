```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7]
y = [20,30,50,30,40,10,90] 
```

```python
df = pd.DataFrame({"days": x, "customer": y})
df
```

```python
sns.lineplot(x="days", y="customer", data=df)
```

```python
# using inbuild dataset
df = sns.load_dataset("tips")
df
```

```python
sns.lineplot(x="day", y="total_bill", data=df)

plt.show()
```

```python
# add hue to categorize male and female
sns.lineplot(x="day", y="total_bill", data=df, hue="sex")
plt.show()
```

```python
# style based on time like is it lunch or dinner
sns.lineplot(x="day", y="total_bill", data=df, hue="sex", style="time")
plt.show()
```

```python
# color palette
sns.lineplot(x="day", y="total_bill", data = df, hue="sex", palette="flare")
plt.show()
```
