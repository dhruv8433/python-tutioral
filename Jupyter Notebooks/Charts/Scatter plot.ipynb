```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("exercise")
df
```

```python
sns.scatterplot(x="time", y="pulse", data=df)
```

```python
sns.scatterplot(x = "time", y = "pulse", hue = "kind", data=df)
```

```python
sns.scatterplot(x = "time", y = "pulse", hue = "kind", data=df, palette="GnBu")
```

```python
sns.scatterplot(x = "time", y = "pulse", data=df, size = "diet")
```

```python
# chaange marker
sns.scatterplot(x = "time", y = "pulse", hue = "kind", data=df, marker="*")
plt.show()
```

```python
# alpha for transperent
sns.scatterplot(x = "time", y = "pulse", hue = "kind", data=df, alpha=0.5)
```
