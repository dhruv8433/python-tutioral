```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("iris")
df
```

```python
# in this graph all the dignoal elements are histrogram because check x and y axis both are similar and 
# if we can see digonal opposite is's mirror of each other just change their axis

sns.pairplot(data = df)
plt.show()
```

```python
# reduce height

sns.pairplot(data = df, height = 2)
plt.show()
```

```python
sns.pairplot(data = df, height = 2, hue = "species")
plt.show()
```

```python
# edge color to circle
sns.pairplot(data = df, height = 2, plot_kws={'edgecolor':'black'})
plt.show()
```
