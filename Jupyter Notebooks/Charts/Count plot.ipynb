```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(x = "sex", data = df)
```

```python
sns.countplot(y = "sex", data = df)
```

```python
# hue
sns.countplot(x = "sex", data = df, hue = "smoker")
```

```python
# palette
sns.countplot(x = "sex", data = df, hue="smoker", palette="PuBuGn")
```

```python
# single color
sns.countplot(x = "sex", data = df, color="hotpink")
```

```python
# saturations
sns.countplot(x = "sex", data = df, color="hotpink", saturation=0.3)
```

```python
# remove color and display only outline with the help of palette
sns.countplot(x = "sex", data = df, facecolor = [0,0,0,0], linewidth = 4, edgecolor = sns.color_palette("GnBu", 2))
```
