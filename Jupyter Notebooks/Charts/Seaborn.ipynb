# Line Plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# set style dark
sns.set(style="dark")

dataset = pd.read_csv("sample.csv") 
print(dataset)
# dataset=sns.load_dataset("fmri")
sns.lineplot(x="id" , y= "rank",data=dataset)
sns.lineplot(x=dataset["id"] , y= dataset["rank"])
# sns.lineplot(x='id', y='rank')
plt.show()
```

# Dist Plot

```python
import numpy as np

# Generate a random univariate type distribution  
ru = np.random.RandomState(10)  
d = ru.normal(size=100)  

sns.histplot(dataset["rank"], kde=True, color="m")  
plt.show()
```
